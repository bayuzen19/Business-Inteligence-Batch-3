3.Buatlah sebuah program yang dapat menghitung jumlah dari n bilangan pertama dari bilangan natural atau bilangan bulat positif. Misal jika user memberikan input n = 10 maka program harus menghitung jumlah dari 1 + 2+ 3+ … + 10

```python
def rumus_jumlah(n):
  hasil = 0
  for suku in range(1, n+1, 1):
      hasil = hasil + suku
  return print("Sn = {}".format(hasil))
```

```python
rumus_jumlah(10)
```

```python
def rumus_segitiga():
  alas = float(input("masukkan nilai alas :"))
  tinggi = float(input("masukkan nilai tinggi : "))
  luas = alas*tinggi/2
  return print(f"luas segitiganya dengan alas {alas} dan tinggi {tinggi} adalah {luas} satuan luas")
```

```python
import math

def heron():
  a = float(input("masukkan nilai a :"))
  b = float(input("masukkan nilai b :"))
  c = float(input("masukkan nilai c :"))
  s = (a+b+c)/2
  rumus_heron = math.sqrt(s*(s-a)*(s-b)*(s-c))
  return print(f"luas segitiga formula heron {rumus_heron}")
```

```python
import numpy as np
def check_gempa():
  magnitudo = float(input("Magnitudo : "))
  if magnitudo >= 0:
    for i in np.arange(0,magnitudo,0.01):
        if i < 2.0:
          return print('Micro earthquake')
        elif i >= 2.0 and i <= 2.9:
          return print('Minor earthquake')
        elif i <= 3.9:
          return print("Minor earthquake")
        elif i <= 4.9:
          return print("Light earthquake")
        elif i <= 5.9:
          return print("Moderate earthquake")
        elif i <= 6.9:
          return print('Strong earthquake')
        elif i <= 7.9:
          return print("Major earthquake")
        elif i <= 8.9:
          return print("Great earthquake")
        elif i <= 9.9:
          return print("Great earthquake")
        else:
          return print("Massive earthquake")
  else:
       return print("Error")
```

```python
#import data
df = pd.read_csv("https://raw.githubusercontent.com/hadimaster65555/dataset_for_teaching/main/dataset/linear_regression_simulation_data/linear-regression.csv")
df = df.T.reset_index().T.reset_index(drop=True)

df.columns  = ["X","Y"]

df["X"] = df['X'].astype(float)
df["Y"] = df["Y"].astype(float)
```

2.Buatlah program untuk menghitung regresi linier menggunakan metode invers matriks <br>
$x = (A^TA)^{-1}A^{T}y$ <br>
dimana: <br>
x adalah koefisien dari model regresi yang ingin dicari <br>
A adalah matriks data <br>
AT adalah transpose dari A <br>
A-1 adalah invers dari matriks A <br>
y adalah variabel target <br>


```python
X = df.X.values
y = df.Y.values

X_mat=np.vstack((np.ones(len(X)), X)).T
beta_hat = np.linalg.inv(X_mat.T.dot(X_mat)).dot(X_mat.T).dot(y)

# predict using coefficients
yhat = X_mat.dot(beta_hat)

# plot data and predictions
plt.scatter(X, y,label="data")
plt.plot(X, yhat, color='red',label="fitted line")
# Menambah teks pada x=0.8 dan y=0.5, berukuran 12 dan berwarna biru
plt.text(0.25,3, r"$\beta_0 = 3.99187633$", fontsize=14, color="b")
# Menambah teks pada x=0.8 dan y=0.3, berukuran 12 dan berwarna blue
plt.text(0.25,2, r"$\beta_1 = 2.98060191$", fontsize=14, color="b")
plt.legend()
```

3. Buatlah program untuk menghitung regresi linier menggunakan method linalg.lstsq dari numpys


```python
# x co-ordinates
x = df.X
A = np.array([x, np.ones(x.shape[0])])

# linearly generated sequence
y = df.Y

# obtaining the parameters of regression line
w = np.linalg.lstsq(A.T, y, rcond=None)[0]
print(w)

line = w[0]*x + w[1]  # regression line
plt.plot(x, line, 'r-',label="fitted line")
plt.plot(x, y, 'o',label='data')
# Menambah teks pada x=0.8 dan y=0.5, berukuran 12 dan berwarna biru
plt.text(0.25,3, r"$\beta_1 = 3.99187633$", fontsize=14, color="b")
# Menambah teks pada x=0.8 dan y=0.3, berukuran 12 dan berwarna blue
plt.text(0.25,2, r"$\beta_0 = 2.98060191$", fontsize=14, color="b")
plt.legend()
plt.show()
```
