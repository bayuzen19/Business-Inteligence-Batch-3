```python
# outcome
y = "achievement_score"
# treatment
T = "intervention"
# covariates
X = ["success_expect", "frst_in_family","gender","ethnicity","school_ethnic_minority"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop("achievement_score",axis=1),df["achievement_score"])
```

```python
# Define X-learner
models = DecisionTreeRegressor(max_depth=4)
propensity_model = RandomForestClassifier(n_estimators=100, max_depth=3)
X_learner = XLearner(models=models, propensity_model=propensity_model)
# Fit learner to data
X_learner.fit(Y = y_train, T = X_train[T], X = X_train[X])
```

```python
# Estimate treatment effects on test data
X_te = X_learner.effect(X_test[X])
```

```python
# measure average treatment
np.mean(X_te)
```

```python
X_learner_df = pd.DataFrame({"model":"X-learner", "treatment_effect": X_te})

#sebaran distribusi
sns.violinplot(x = "model", y = "treatment_effect", data = X_learner_df);
```
